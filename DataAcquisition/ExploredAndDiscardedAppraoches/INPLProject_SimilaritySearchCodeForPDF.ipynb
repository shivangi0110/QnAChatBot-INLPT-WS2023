```python
from langchain.document_loaders import TextLoader
from langchain.document_loaders import UnstructuredPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
# from langchain.embeddings import OpenAIEmbeddings
# from dotenv import load_dotenv

# load_dotenv()
```

```python
loader = UnstructuredPDFLoader('/content/NLP_Project/CELEX_21975A1201(01)_EN_TXT.pdf', encoding='utf-8')
documents = loader.load()

print(documents)  # prints the document objects
print(len(documents))  # 1 - we've only read one file/document into the loader
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
texts = text_splitter.split_documents(documents)

print(texts)
print(len(texts))
```

```python
print(texts[0])
```

```python
import spacy

# Load the spaCy model (you can choose a different model based on your language and requirements)
nlp = spacy.load("en_core_web_sm")

# Function to obtain word embeddings for a given text
def get_embeddings(text):
    # Process the text with spaCy
    doc = nlp(text)

    # Get the average word embedding for the document
    # You can also choose other strategies like summing or using the vector of a specific token
    if len(doc) > 0:
        return doc.vector
    else:
        return None

# Apply the get_embeddings function to each text chunk
embeddings = [get_embeddings(text.page_content) for text in texts[:5]]


# Print the embeddings
print(embeddings)
print(len(embeddings))
```

```python
import numpy as np
import faiss

# Convert the list of embeddings to a 2D numpy array
embeddings_array = np.array(embeddings)

# Check if there are valid embeddings
if embeddings_array.shape[0] == 0:
    print("No valid embeddings found.")
else:
    # Print the embeddings
    print(embeddings_array)
    print(embeddings_array.shape)

    # Initialize Faiss index
    index = faiss.IndexFlatL2(embeddings_array.shape[1])

    # Add embeddings to the index
    index.add(embeddings_array)
```

```python
import numpy as np
import faiss

import spacy

# Load pre-trained spaCy model (you'll need to download a spaCy model)
# For example, 'en_core_web_md' includes 300-dimensional word vectors
nlp = spacy.load("en_core_web_sm")

# Function to get the spaCy embedding for a given text
def get_embedding_from_model(text):

    doc = nlp(text)

    # Use the vector for the entire document
    if len(doc) > 0:
      return doc.vector
    else:
      return None


# Function to perform similarity search
def search(query_embedding_array, index, embeddings_array, k=5):

    # Search for the nearest neighbors
    _, neighbors = index.search(query_embedding_array, k)

    # print(neighbors[0])

    return neighbors[0]

# Provide a query text
query_text = "How do the International Atomic Energy Agency and the European Atomic Energy Community agree to cooperate, and what is their approach to harmonizing efforts?"

# Get the embedding for the query text
query_embedding = get_embedding_from_model(query_text)

# Convert the query embedding to a numpy array
query_embedding_array = np.array([query_embedding])

# Perform similarity search for the query
distances, result_indices = index.search(query_embedding_array, k=5)

#print(result_indices)

# Display the results
print("Query Text:")
print(query_text)

print("\nSimilar Documents:")

# 'result_indices' is the list of indices obtained from the similarity search
for distance, idx in zip(distances[0], result_indices[0]):
    print(f"\nDocument Index: {idx}")
    print("Similarity Score:", 1 - distance)
    print("\n")
    print(texts[idx].page_content)



```
