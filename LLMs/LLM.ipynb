```python
from torch import cuda, bfloat16
import os
import transformers
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
model_id = 'meta-llama/Llama-2-13b-chat-hf'
```

```python
embedding_model = 'sentence-transformers/all-MiniLM-L6-v2'
device = 'cuda:0'

embed_model = HuggingFaceEmbeddings(
    model_name=embedding_model,
    model_kwargs={'device': device},
    encode_kwargs={'device': device, 'batch_size': 32}
)
```

```python
from transformers import LlamaTokenizer
hf_auth = os.environ.get('HF_AUTH')
tokenizer = LlamaTokenizer.from_pretrained("meta-llama/Llama-2-13b-chat-hf",use_auth_token=hf_auth)
```

```python
bitsAndBites_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=bfloat16
)
```

```python
model_config = transformers.AutoConfig.from_pretrained(
    model_id,
    use_auth_token=os.environ.get('HF_AUTH')
)
```

```python
model = transformers.AutoModelForCausalLM.from_pretrained(
    model_id,
    trust_remote_code=True,
    config=model_config,
    quantization_config=bitsAndBites_config,
    device_map='auto',
    token=os.environ.get('HF_AUTH')
)
```

```python
generate_text = transformers.pipeline(
    model=model, tokenizer=tokenizer,
    return_full_text=True,
    task='text-generation',

    temperature=0.01,
    max_new_tokens=512,
    repetition_penalty=1.1
)
```

```python
from ipywidgets import interact, widgets
from IPython.display import display


def generate_answer_from_question(question):

    return retrieve_answers(question)

# Function to handle user input and display the generated answer
def handle_input(questions):
    questions_list = questions.split('\n')
    results = []  # List to store the results
    for question in questions_list:
        answer = generate_answer_from_question(question)
        results.append((question, answer["result"]))  # Store the query and answer
    return results

# Create text input widget for user input
input_widget = widgets.Textarea(
    value='',
    placeholder='Type your questions here, each separated by a new line...',
    description='Input:',
    disabled=False
)

# Create button widget to trigger query processing
button_widget = widgets.Button(description="Submit")

# Define callback function when button is clicked
def on_button_clicked(b):
    user_input = input_widget.value
    results = handle_input(user_input)
    # Display all results
    for question, answer in results:
        print("Question:", question)
        print("Generated Answer:", answer)
        print()

button_widget.on_click(on_button_clicked)

# Display the widgets
display(input_widget)
display(button_widget)
```
